```python
subject = 'sub-01'
num_run = 5

# General import 
import os
import platform
import numpy as np
import ipdb
import scipy.io
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.ticker import FormatStrFormatter

%matplotlib inline

# Get eyelink data
if platform.system() == 'Darwin':
    edf2asc_dir = '/Applications/Eyelink/EDF_Access_API/Example';
    end_file = '';
    full_exp_dir = '/Users/martin/Dropbox/Experiments/pMFexp'
elif platform.system() == 'Windows':
    edf2asc_dir = 'C:/Users/invibe/Experiments/pMFexp/stats/';
    end_file ='.exe';
    full_exp_dir = 'C:/Users/invibe/Experiments/pMFexp/'

time_last_run_eye = 0;
fix_accuracy = [];
eye_data_runs = [];
task1_txt = 'EyeMov'

file_dir = '{exp_dir}/data/{sub}'.format(exp_dir = full_exp_dir, 
                                         sub = subject)

list_filename = ['{sub}_task-{task}_run-01'.format(sub = subject, task = task1_txt),
                 '{sub}_task-{task}_run-02'.format(sub = subject, task = task1_txt),
                 '{sub}_task-{task}_run-03'.format(sub = subject, task = task1_txt),
                 '{sub}_task-{task}_run-04'.format(sub = subject, task = task1_txt),
                 '{sub}_task-{task}_run-05'.format(sub = subject, task = task1_txt),
                 '{sub}_task-{task}_run-06'.format(sub = subject, task = task1_txt),
                 '{sub}_task-{task}_run-07'.format(sub = subject, task = task1_txt),
                 '{sub}_task-{task}_run-08'.format(sub = subject, task = task1_txt),
                 '{sub}_task-{task}_run-09'.format(sub = subject, task = task1_txt),
                 '{sub}_task-{task}_run-10'.format(sub = subject, task = task1_txt)]
```

```python
time_last_run_eye   =   0;
time_start_eye = np.zeros((1,num_run))
time_end_eye = np.zeros((1,num_run))
for t_run in np.arange(0,num_run,1):
    
    edf_filename = '{file_dir}/func/{filename}_eyeData'.format(file_dir = file_dir,filename = list_filename[t_run]);
    mat_filename = '{file_dir}/add/{filename}_matFile.mat'.format(file_dir = file_dir,filename = list_filename[t_run]);
    # get .msg and .dat file
    if not os.path.exists('{}.dat'.format(edf_filename)) or not os.path.exists('{}.msg'.format(edf_filename)):
        os.system('{edf2asc_dir}/edf2asc{end_file} {edf_filename}.edf -e -y'.format(edf2asc_dir = edf2asc_dir,
                                                                                    end_file = end_file,
                                                                                    edf_filename = edf_filename))
        os.rename('{}.asc'.format(edf_filename),'{}.msg'.format(edf_filename))
        
        os.system('{edf2asc_dir}/edf2asc{end_file} {edf_filename}.edf -s -miss -1.0 -y'.format( edf2asc_dir = edf2asc_dir,
                                                                                                end_file = end_file,
                                                                                                edf_filename = edf_filename))
        os.rename('{}.asc'.format(edf_filename),'{}.dat'.format(edf_filename))
    
    # get first and last time pf each run
    msgfid = open('{}.msg'.format(edf_filename))    
    first_last_time = False
    first_time = False
    last_time = False    
    while not first_last_time:
        line_read = msgfid.readline()
        if not line_read == '':
            la = line_read.split()
            
            if len(la) > 6:
                if la[2] == 'sequence' and la[3]=='1' and la[4]=='started' and not first_time: 
                    time_start_eye[0,t_run] = float(la[1])
                    first_time = True
                    
                if la[2] == 'sequence' and la[3]=='9' and la[4]=='stopped' and not last_time: 
                    time_end_eye[0,t_run] = float(la[1])
                    last_time = True
                    
            if first_time == True and last_time == True:
                first_last_time = True
                msgfid.close();
        
    # load eye coord data
    eye_dat = np.genfromtxt('{}.dat'.format(edf_filename),usecols=(0, 1, 2))
    eye_data_run = eye_dat[np.logical_and(eye_dat[:,0]>=time_start_eye[0,t_run],eye_dat[:,0]<=time_end_eye[0,t_run])]
    # col 1 = time
    # col 2 = eye x coord
    # col 3 = eye y coord
    
    # compute time relative to start of trial and across blocks
    eye_data_run[:,0] =  (eye_data_run[:,0]-time_start_eye[0,t_run])+time_last_run_eye;
    time_last_run_eye = eye_data_run[-1,0];
    if t_run == 0:
        eye_data_runs = eye_data_run
    else:
        eye_data_runs = np.vstack((eye_data_runs,eye_data_run))
        
    # remove msg and dat 
    os.remove('{}.msg'.format(edf_filename))
    os.remove('{}.dat'.format(edf_filename))
```

```python
# delete blink time
blinkNum = 0;
blink_start = False;
for tTime in np.arange(0,eye_data_runs.shape[0],1):
    if not blink_start:
        if eye_data_runs[tTime,1] == -1:
            blinkNum += 1
            timeBlinkOnset = eye_data_runs[tTime,0]
            blink_start = True
            if blinkNum == 1:
                blink_onset_offset = np.matrix([timeBlinkOnset,np.nan])
            else:
                blink_onset_offset = np.vstack((blink_onset_offset,[timeBlinkOnset,np.nan]))

    if blink_start:
        if eye_data_runs[tTime,1] != -1:
            timeBlinkOffset = eye_data_runs[tTime,0]
            blink_start = 0
            blink_onset_offset[blinkNum-1,1] = timeBlinkOffset

# nan record around detected blinks
befDurBlink = 300;   # duration before blink
aftDurBlink = 300;   # duration after blink
eye_data_runs_no_blink = np.copy(eye_data_runs)
for tBlink in np.arange(0,blinkNum,1):
    blink_onset_offset[tBlink,0] = blink_onset_offset[tBlink,0]-befDurBlink
    blink_onset_offset[tBlink,1] = blink_onset_offset[tBlink,1]+aftDurBlink
    eye_data_runs_no_blink[np.logical_and(eye_data_runs_no_blink[:,0] >= blink_onset_offset[tBlink,0],eye_data_runs_no_blink[:,0] <= blink_onset_offset[tBlink,1]),1] = np.nan
    eye_data_runs_no_blink[np.logical_and(eye_data_runs_no_blink[:,0] >= blink_onset_offset[tBlink,0],eye_data_runs_no_blink[:,0] <= blink_onset_offset[tBlink,1]),2] = np.nan
```

```python
# compute in time percentage between start and end
time_prc_eye = eye_data_runs[:,0]/np.sum(time_end_eye - time_start_eye)
```

```python
# put eye coordinates in deg from center (flip y axis)
matfile = scipy.io.loadmat(mat_filename)
scr_sizeX = matfile['config']['scr'][0,0]['scr_sizeX'][0][0][0][0]
scr_sizeY = matfile['config']['scr'][0,0]['scr_sizeY'][0][0][0][0]
screen_size = np.array([scr_sizeX,scr_sizeY])
ppd = matfile['config']['const'][0,0]['ppd'][0][0][0][0]

eye_data_runs[:,1] = (eye_data_runs[:,1] - (screen_size[0]/2))/ppd;
eye_data_runs[:,2] = (eye_data_runs[:,2] - (screen_size[1]/2))/ppd;
eye_data_runs_no_blink[:,1] = (eye_data_runs_no_blink[:,1] - (screen_size[0]/2))/ppd;
eye_data_runs_no_blink[:,2] = (eye_data_runs_no_blink[:,2] - (screen_size[1]/2))/ppd;
```

```python
# Define figure
title_font = {'loc':'left', 'fontsize':14, 'fontweight':'bold'}
axis_label_font = {'fontsize':14}
bg_col = (0.9, 0.9, 0.9)
axis_width = 0.75
line_width_corr = 1.5
rads = [2.5,5,7.5,10]

# Horizontal eye trace
screen_val =  12.5
ymin1,ymax1,y_tick_num1 = -screen_val,screen_val,11
xmin1,xmax1,x_tick_num1 = 0,1,num_run+1
y_tick1 = np.linspace(ymin1,ymax1,y_tick_num1)
x_tick1 = np.linspace(xmin1,xmax1,x_tick_num1)

# Vertical eye trace
ymin2,ymax2,y_tick_num2 = -screen_val,screen_val,11
xmin2,xmax2,x_tick_num2 = 0,1,num_run+1
y_tick2 = np.linspace(ymin2,ymax2,y_tick_num2)
x_tick2 = np.linspace(xmin2,xmax2,x_tick_num2)

try: os.makedirs('{file_dir}/add/figures/'.format(file_dir = file_dir))
except: pass
```

```python
for run_plot in np.arange(0,num_run,1):
    
    
    fig = plt.figure(figsize = (15, 7))
    gridspec.GridSpec(2,8)
    logic_min = run_plot/num_run
    logic_max = (run_plot+1)/num_run
    logic_run = np.logical_and(time_prc_eye>=logic_min,time_prc_eye<=logic_max)
    
    xmin1,xmax1,x_tick_num1 = logic_min,logic_max,5
    xmin2,xmax2,x_tick_num2 = logic_min,logic_max,5
    x_tick1 = np.linspace(xmin1,xmax1,x_tick_num1)
    x_tick2 = np.linspace(xmin2,xmax2,x_tick_num2)

    # Horizontal eye trace
    ax1 = plt.subplot2grid((2,8),(0,0),rowspan= 1, colspan = 4)
    ax1.set_ylabel('Hor. coord. (dva)',axis_label_font,labelpad = 0)
    ax1.set_ylim(bottom = ymin1, top = ymax1)
    ax1.set_yticks(y_tick1)
    ax1.set_xlabel('Time (%)',axis_label_font,labelpad = 10)
    ax1.set_xlim(left = xmin1, right = xmax1)
    ax1.set_xticks(x_tick1)
    ax1.set_facecolor(bg_col)
    ax1.set_title('Horizontal eye position',**title_font)
    ax1.xaxis.set_major_formatter(FormatStrFormatter('%.2g'))
    for rad in rads:
        ax1.plot(time_prc_eye[logic_run],time_prc_eye[logic_run]*0+rad, color = [1,1,1], linewidth = axis_width*2)
        ax1.plot(time_prc_eye[logic_run],time_prc_eye[logic_run]*0-rad, color = [1,1,1], linewidth = axis_width*2)
    
    ax1.plot(time_prc_eye[logic_run],eye_data_runs_no_blink[logic_run,1],color = [0.5,0.5,0.5],linewidth = axis_width*1.5)
    
    # Vertical eye trace
    ax2 = plt.subplot2grid((2,8),(1,0),rowspan= 1, colspan = 4)
    ax2.set_ylabel('Ver. coord. (dva)',axis_label_font, labelpad = 0)
    ax2.set_ylim(bottom = ymin2, top = ymax2)
    ax2.set_yticks(y_tick2)
    ax2.set_xlabel('Time (%)',axis_label_font, labelpad = 10)
    ax2.set_xlim(left = xmin2, right = xmax2)
    ax2.set_xticks(x_tick2)
    ax2.set_facecolor(bg_col)
    ax2.set_title('Vertical eye position',**title_font)
    ax2.xaxis.set_major_formatter(FormatStrFormatter('%.2g'))
    for rad in rads:
        ax2.plot(time_prc_eye[logic_run],time_prc_eye[logic_run]*0+rad, color = [1,1,1], linewidth = axis_width*2)
        ax2.plot(time_prc_eye[logic_run],time_prc_eye[logic_run]*0-rad, color = [1,1,1], linewidth = axis_width*2)

    ax2.plot(time_prc_eye[logic_run],eye_data_runs_no_blink[logic_run,2],color = [0.5,0.5,0.5],linewidth = axis_width*1.5)

    # Screen eye trace
    ax3 = plt.subplot2grid((2,8),(0,4),rowspan= 2, colspan = 4)
    ax3.set_xlabel('Horizontal coordinates (dva)', axis_label_font, labelpad = 10)
    ax3.set_ylabel('Vertical coordinates (dva)', axis_label_font, labelpad = 0)
    ax3.set_xlim(left = ymin1, right = ymax1)
    ax3.set_xticks(y_tick1)
    ax3.set_ylim(bottom = ymin2, top = ymax2)
    ax3.set_yticks(y_tick2)
    ax3.set_facecolor(bg_col)
    ax3.set_title('Screen view',**title_font)
    ax3.set_aspect('equal')
    
    theta = np.linspace(0, 2*np.pi, 100)
    for rad in rads:
        ax3.plot(rad*np.cos(theta), rad*np.sin(theta),color = [1,1,1],linewidth = axis_width*3)
        
    ax3.plot(eye_data_runs_no_blink[logic_run,1],eye_data_runs_no_blink[logic_run,2],color = [0.5,0.5,0.5],linewidth = axis_width*1.5)
    ax3.text(8, -10, 'Run {run_plot}'.format(run_plot = run_plot+1), horizontalalignment = 'left', verticalalignment = 'center', fontsize = 14)

    plt.subplots_adjust(wspace = 1.4,hspace = 0.4)
    plt.savefig("{file_dir}/add/figures/{filename}_eyeData_eyetraces.png".format(file_dir = file_dir, filename = list_filename[run_plot]),facecolor='w')
    
```
